```python
# Example of a simple GET request using requests library

import requests

response = requests.get('https://api.github.com/events')
if response.status_code == 200:
    print("Success!")
    print(response.url)  # This will print the URL of the response
    print(response.headers)  # This will print the headers of the response (e.g., content type, date, etc.) metadata
    print(type(response.headers))  # This will print the type of headers, which is a dictionary

    content_type = response.headers.get('Content-Type') # This will get the content type from the headers
    print("Content-Type:", content_type)
else:
    print("Failed to retrieve data:", response.status_code)
```

```python
import requests
import json

response = requests.get('https://api.github.com/events')
if response.status_code == 200:
    print("Success!")
    if 'application/json;' in response.headers.get('Content-Type', ''):
        data = response.json()
        print(data)
        print(type(data))
    print(response.text)
else:
    print("Error while fetching data:", response.status_code)
```

```python
# send Post request
import requests
import json
url = 'https://httpbin.org/post'
# Data to be in the form
data = {'username': 'supunmadhuranga', 'password' : 'your_password'}
response = requests.post(url, data=data)
if response.status_code == 200:
    print("Post request successful!")
    print(response.text)  # This will print the content of the response as a string
    print(response.url)  # This will print the URL of the response
else:
    print("Post request failed with status code:", response.status_code)
```

```python
# send Get request with parameters

import requests

url = 'https://jsonplaceholder.typicode.com/posts'

# Parameters to be sent in the URL
# This will be appended to the URL as query parameters
params = {'userId': 1 , 'id': 1}

response = requests.get(url, params=params)

print(response.url)  # This will print the URL with parameters
print(response.text)  # This will print the content of the response as a string
```

```python
# send Get request with authentication

import requests
import json
url = 'https://httpbin.org/post'

username = 'supunmadhuranga'
password = 'your_password'

response = requests.get(url , auth=(username, password)) # Send GET request with authentication

if response.status_code == 200: 
    print("Authentication successful!")
    print(response.text)  # This will print the content of the response as a string
else:
    print("Authentication failed with status code:", response.status_code)
```

```python
from bs4 import BeautifulSoup

html_doc = """<html><head><title>The Dormouse's story</title></head>
<body>
<p class="title"><b>The Dormouse's story</b></p>
<p class="story">Once upon a time there were three little sisters; and their names were
<a href="http://example.com/elsie&quot; class="sister" id="link1">Elsie</a>,
<a href="http://example.com/lacie&quot; class="sister" id="link2">Lacie</a> and
<a href="http://example.com/tillie&quot; class="sister" id="link3">Tillie</a>;
and they lived at the bottom of a well.</p>
<p class="story">...</p>"""

soup = BeautifulSoup(html_doc, 'html.parser')
print(soup.title)  # This will print the title tag
print(soup.p) # This will print the first paragraph tag
print(soup.a) # This will print the first anchor tag

print(soup.find('p')) # This will find the first paragraph tag
print(soup.find_all('p'))  # This will find all paragraph tags


print(soup.title.string)  # This will print the text inside the title tag
```

```python
import requests
import json
from bs4 import BeautifulSoup

url = "http://example.com"
response = requests.get(url)
if response.status_code == 200:
    print("Successfull")
    soup = BeautifulSoup(response.text, 'html.parser')
    
    h1_tag = soup.find('h1')
    if h1_tag:
        print(h1_tag.string) # This will print the text inside the h1 tag
        # print(h1_tag.text)  # This will print the text inside the h1 tag
    else:
        print("No h1 tag found in the response")
else:
    print("Failed to retrieve the page:", response.status_code)
```
